```python
import numpy as np #manipular arrays
import matplotlib.pyplot as plt # visualizar dados 
from sklearn.linear_model import LinearRegression #criar modelo de regressão
from sklearn.metrics import mean_squared_error, r2_score  # avaliação de modelo 
```

```python
x = np.array([6, 8, 10, 14, 18]) #armazenando os diametros de pizza em polegas em arrays 
y = np.array([7, 9, 13, 17.5, 18]) # armazenando os valores da pizza
```

```python
plt.scatter(x, y) # criação de grafico de disperão para visualizar a relação entre as 2 variaveis 
plt.xlabel('Diâmetro em polegadas') 
plt.ylabel ('Preço em reais')
plt.title("Preço da pizza em razão do diametro")
plt.show()
```

```python
model = LinearRegression() 
x = x.reshape(-1, 1)
model.fit(x, y) # .fit serve para invocar o aprendizado da maquina e a busca da relação entre a variavel x e y 
```

```python
y_pred = model.predict(x)
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y , y_pred)
print(mse)
print(r2)
```

```python
x_new = np.array([20])
y_new = model.predict(x_new.reshape(-1, 1))
```
